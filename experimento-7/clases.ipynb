```python
import os
from collections import Counter

def contar_clases(carpeta_labels, nombres_clases):
    conteo = Counter()
    
    for archivo in os.listdir(carpeta_labels):
        if archivo.endswith(".txt"):
            with open(os.path.join(carpeta_labels, archivo), "r") as f:
                for linea in f:
                    if linea.strip():
                        clase = int(linea.split()[0])
                        conteo[clase] += 1
    
    print("=== Balance de clases ===")
    total = sum(conteo.values())
    for clase_id, cantidad in sorted(conteo.items()):
        nombre = nombres_clases[clase_id]
        porcentaje = (cantidad / total) * 100
        print(f"  Clase {clase_id} ({nombre}): {cantidad} instancias ({porcentaje:.1f}%)")
    
    return conteo

# Corre esto antes de entrenar
conteo = contar_clases(
    carpeta_labels="dataset/train/labels",
    nombres_clases=["normal", "robo"]
)
```
