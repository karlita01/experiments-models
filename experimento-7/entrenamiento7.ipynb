```python
from ultralytics import YOLO

model = YOLO("yolov8s.pt")

results = model.train(
    data="dataset/data.yaml",

    # Entrenamiento
    epochs=100,
    patience=20,
    imgsz=640,
    batch=32,
    workers=6,

    # Optimizador
    optimizer="AdamW",
    lr0=0.0005,
    lrf=0.01,
    weight_decay=0.0005,
    warmup_epochs=3,

    # Augmentación
    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.4,
    fliplr=0.5,
    flipud=0.2,
    mosaic=1.0,
    mixup=0.15,
    degrees=10.0,
    scale=0.5,
    translate=0.1,

    # GPU
    device=0,
    amp=True,

    # Guardado
    name="robo_v2",
    project="runs/train",
    save=True,
    save_period=10,
    plots=True,
)

print("✅ Entrenamiento terminado")
print(f"Mejor modelo en: {results.save_dir}/weights/best.pt")
```

```python
import cv2
from ultralytics import YOLO

model = YOLO("runs/detect/runs/train/robo_v2/weights/best.pt")

# Colores por clase: normal=verde, robo=rojo
COLORES = {
    0: (0, 255, 0),    # normal - verde
    1: (0, 0, 255),    # robo - rojo
}

cap = cv2.VideoCapture("video2.mp4")  # 0 para webcam

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    results = model.predict(frame, conf=0.4, verbose=False)

    for r in results:
        for box in r.boxes:
            clase_id = int(box.cls[0])
            confianza = float(box.conf[0])
            nombre = model.names[clase_id]
            color = COLORES.get(clase_id, (255, 255, 255))

            # Coordenadas del bounding box
            x1, y1, x2, y2 = map(int, box.xyxy[0])

            # Dibujar caja
            cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)

            # Etiqueta con clase y confianza
            label = f"{nombre} {confianza:.2f}"
            cv2.putText(frame, label, (x1, y1 - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.6, color, 2)

    # Mostrar frame
    cv2.imshow("Deteccion de Actividad Sospechosa", frame)

    # Presiona 'q' para salir
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
from ultralytics import YOLO

# Modelo 1: detecta personas individualmente (preentrenado COCO)
detector = YOLO("yolov8s.pt")

# Modelo 2: tu clasificador robo/normal
clasificador = YOLO("runs/detect/runs/train/robo_v2/weights/best.pt")

COLORES = {
    0: (0, 255, 0),    # normal - verde
    1: (0, 0, 255),    # robo - rojo
}

cap = cv2.VideoCapture("video2.mp4")

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Paso 1: detectar y trackear personas
    tracks = detector.track(frame, conf=0.4, persist=True,
                            classes=[0], verbose=False)  # clase 0 = persona en COCO

    for r in tracks:
        if r.boxes is None:
            continue

        for box in r.boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            track_id = int(box.id[0]) if box.id is not None else -1

            # Paso 2: recortar la persona y clasificar actividad
            crop = frame[y1:y2, x1:x2]
            if crop.size == 0:
                continue

            result_cls = clasificador.predict(crop, conf=0.3, verbose=False)

            # Obtener clase con mayor confianza
            clase_id = 0  # default normal
            confianza = 0.0
            if result_cls[0].boxes and len(result_cls[0].boxes) > 0:
                clase_id = int(result_cls[0].boxes[0].cls[0])
                confianza = float(result_cls[0].boxes[0].conf[0])

            nombre = clasificador.names[clase_id]
            color = COLORES.get(clase_id, (255, 255, 255))

            # Dibujar caja y etiqueta
            cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)
            label = f"ID:{track_id} {nombre} {confianza:.2f}"
            cv2.putText(frame, label, (x1, y1 - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.6, color, 2)

    cv2.imshow("Deteccion de Actividad Sospechosa", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
