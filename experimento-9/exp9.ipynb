```python
import os
from collections import Counter, defaultdict

DATASET_PATH = "dataset"

splits = {
    "train": os.path.join(DATASET_PATH, "train", "labels"),
    "valid": os.path.join(DATASET_PATH, "valid", "labels"),
    "test": os.path.join(DATASET_PATH, "test", "labels"),
}

class_names = {
    0: "normal",
    1: "robo"
}

total_instances = Counter()
total_images_with_class = defaultdict(set)
label_format_counter = Counter()
bad_files = []

for split, labels_path in splits.items():
    print(f"\n===== Revisando {split} =====")
    
    split_instances = Counter()
    split_images_with_class = defaultdict(set)
    txt_files = [f for f in os.listdir(labels_path) if f.endswith(".txt")]
    
    for txt_file in txt_files:
        file_path = os.path.join(labels_path, txt_file)
        
        with open(file_path, "r", encoding="utf-8") as f:
            lines = [line.strip() for line in f.readlines() if line.strip()]
        
        for line in lines:
            parts = line.split()
            
            try:
                class_id = int(float(parts[0]))
            except:
                bad_files.append((split, txt_file, "clase inválida", line))
                continue
            
            split_instances[class_id] += 1
            total_instances[class_id] += 1
            
            split_images_with_class[class_id].add(txt_file)
            total_images_with_class[class_id].add((split, txt_file))
            
            if len(parts) == 5:
                label_format_counter["deteccion_box_5_valores"] += 1
            elif len(parts) > 5:
                label_format_counter["segmentacion_o_poligono_mas_de_5_valores"] += 1
                bad_files.append((split, txt_file, f"{len(parts)} valores", line[:120]))
            else:
                label_format_counter["label_incompleto_menos_de_5_valores"] += 1
                bad_files.append((split, txt_file, f"{len(parts)} valores", line))
    
    print("Imágenes con labels:", len(txt_files))
    
    print("\nInstancias por clase:")
    for class_id, count in split_instances.items():
        name = class_names.get(class_id, f"clase_{class_id}")
        print(f"  {class_id} - {name}: {count}")
    
    print("\nImágenes por clase:")
    for class_id, files in split_images_with_class.items():
        name = class_names.get(class_id, f"clase_{class_id}")
        print(f"  {class_id} - {name}: {len(files)}")

print("\n===== RESUMEN GENERAL =====")

print("\nTotal de instancias por clase:")
for class_id, count in total_instances.items():
    name = class_names.get(class_id, f"clase_{class_id}")
    print(f"  {class_id} - {name}: {count}")

print("\nTotal de imágenes por clase:")
for class_id, files in total_images_with_class.items():
    name = class_names.get(class_id, f"clase_{class_id}")
    print(f"  {class_id} - {name}: {len(files)}")

print("\nFormato de etiquetas:")
for fmt, count in label_format_counter.items():
    print(f"  {fmt}: {count}")

print("\nArchivos con posible problema:", len(bad_files))

print("\nPrimeros 20 problemas encontrados:")
for item in bad_files[:20]:
    print(item)
```

```python
import os
import shutil
from pathlib import Path

ORIGINAL_DATASET = Path("dataset")
CLEAN_DATASET = Path("dataset_clean")

splits = ["train", "valid", "test"]

# 1. Copiar todo el dataset original
if CLEAN_DATASET.exists():
    shutil.rmtree(CLEAN_DATASET)

shutil.copytree(ORIGINAL_DATASET, CLEAN_DATASET)

def polygon_to_bbox(coords):
    """
    Convierte coordenadas x1 y1 x2 y2 x3 y3... a bbox YOLO:
    x_center y_center width height
    """
    xs = coords[0::2]
    ys = coords[1::2]

    x_min = min(xs)
    x_max = max(xs)
    y_min = min(ys)
    y_max = max(ys)

    x_center = (x_min + x_max) / 2
    y_center = (y_min + y_max) / 2
    width = x_max - x_min
    height = y_max - y_min

    return x_center, y_center, width, height

problemas = []

for split in splits:
    labels_dir = CLEAN_DATASET / split / "labels"

    for txt_file in labels_dir.glob("*.txt"):
        nuevas_lineas = []

        with open(txt_file, "r", encoding="utf-8") as f:
            lines = f.readlines()

        for line in lines:
            line = line.strip()

            if not line:
                continue

            parts = line.split()

            try:
                class_id = int(float(parts[0]))
                values = list(map(float, parts[1:]))
            except:
                problemas.append((str(txt_file), line))
                continue

            # Caso correcto: YOLO detect
            if len(parts) == 5:
                x, y, w, h = values
                nuevas_lineas.append(f"{class_id} {x:.6f} {y:.6f} {w:.6f} {h:.6f}")

            # Caso segmentación: más de 5 valores
            elif len(parts) > 5:
                if len(values) % 2 == 0:
                    x, y, w, h = polygon_to_bbox(values)

                    # Evitar cajas inválidas
                    x = max(0, min(1, x))
                    y = max(0, min(1, y))
                    w = max(0, min(1, w))
                    h = max(0, min(1, h))

                    nuevas_lineas.append(f"{class_id} {x:.6f} {y:.6f} {w:.6f} {h:.6f}")
                else:
                    problemas.append((str(txt_file), line))

            else:
                problemas.append((str(txt_file), line))

        with open(txt_file, "w", encoding="utf-8") as f:
            f.write("\n".join(nuevas_lineas))

print("Dataset limpio creado en:", CLEAN_DATASET)
print("Problemas encontrados:", len(problemas))

for p in problemas[:10]:
    print(p)
```

```python
import os

labels_dirs = [
    "dataset_clean/train/labels",
    "dataset_clean/valid/labels",
    "dataset_clean/test/labels"
]

box_count = 0
segment_count = 0
bad_count = 0

for labels_dir in labels_dirs:
    for file in os.listdir(labels_dir):
        if file.endswith(".txt"):
            path = os.path.join(labels_dir, file)

            with open(path, "r", encoding="utf-8") as f:
                for line in f:
                    parts = line.strip().split()

                    if not parts:
                        continue

                    if len(parts) == 5:
                        box_count += 1
                    elif len(parts) > 5:
                        segment_count += 1
                    else:
                        bad_count += 1

print("Labels correctos de detección:", box_count)
print("Labels con más de 5 valores:", segment_count)
print("Labels incompletos:", bad_count)
```

```python
from ultralytics import YOLO

model = YOLO("yolov8s.pt")

results = model.train(
    data="dataset_clean/data.yaml",
    epochs=20,
    patience=10,
    imgsz=640,
    batch=16,
    workers=2,
    device=0,
    amp=True,
    name="robo_yolov8s_clean_prueba20",
    project="runs/train",
    save=True,
    plots=True
)
```
